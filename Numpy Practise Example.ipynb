```python
# Now using numpy and matplotlib let us try to plot sin graph
import numpy as np
from matplotlib import pyplot as pt
```

```python
a=np.arange(0,3*np.pi,0.5)
print(a)
b=np.sin(a)
print(b)
```

```python
pt.plot(a,b)
pt.show()
```

```python
# lets create 6x6 array  2d array, and place 1 and 0  in alternative across the diagonals
a=np.zeros((6,6),dtype=int)
# print(a)
a[0::2,0::2]=1
a[1::2,1::2]=1
print(a)
```

```python
# Find the totalnumber and locations of missing values in the array
a=np.random.rand(10,10)
print(a)
```

```python
a[np.random.randint(10,size=5),np.random.randint(10,size=5)]=np.nan
print(a)
```

```python
print("Number of nan values in the array --\n",np.isnan(a).sum())
print("The index of the nan numbers :\n",np.argwhere(np.isnan(a)))
```

```python
nan=np.where(np.isnan(a))
a[nan]=0
print(a)
```
